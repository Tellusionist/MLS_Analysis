# Exploratory Analysis of the MLS

```python
import requests
import json
import pandas as pd
```

```python
amer_soccer_mls_url = "https://app.americansocceranalysis.com/api/v1/mls/"
url = amer_soccer_mls_url + 'teams'
r = requests.get(url)
```

```python
teams = pd.read_json(json.dumps(r.json()))
teams.head()
```

```python
url = amer_soccer_mls_url + 'games'
params = {'team_id':'WBLMvYAQxe'}
r = requests.get(url,params)
```

```python
len(r.json())
```

```python
pd.read_json(json.dumps(r.json()), convert_dates=False)
```
